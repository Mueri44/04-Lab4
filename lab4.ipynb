```python
import cv2
import matplotlib.pyplot as plt
# Load the image in grayscale
img = cv2.imread('cameraman.png', cv2.IMREAD_GRAYSCALE)
```

```python
# Set the threshold values
threshold_values = [0, 50, 100, 150, 200]

# Apply thresholding with different threshold values
for threshold_value in threshold_values:
    ret, thresh = cv2.threshold(img, threshold_value, 255, cv2.THRESH_BINARY)
    plt.figure(figsize=(4,4))
    plt.imshow(thresh, cmap='gray')
    plt.title(f'Thresholded Image ({threshold_value})')
    plt.axis('off')
    plt.show()
```

```python
# Display the original image
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
# Task#2: Histogram Processing

# Step#1: Load necessary libraries
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

from skimage import data, img_as_float
from skimage import exposure

# Step#2: Load moon image
img = data.moon()
```

```python
# Contrast stretching
p2, p98 = np.percentile(img, (3, 80))
img_rescale = exposure.rescale_intensity(img, in_range=(p2, p98))
```

```python
# Step#4: Display the image with its histogram of (step#2)  and (step#3)

# Task 2: Histogram Equalization
img_eq = exposure.equalize_hist(img)

# Display all results
plt.figure(figsize=(15,10))

# Original 
plt.subplot(3,2,1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(3,2,2)
plt.hist(img.flat, bins=256, range=(0,255))
plt.title('Original Histogram')


# Rescale (Task 1)
plt.subplot(3,2,3)
plt.imshow(img_rescale, cmap='gray')
plt.title('Contrast Stretch (3%,80%)')
plt.axis('off')
plt.subplot(3,2,4)
plt.hist(img_rescale.flat, bins=256, range=(0,255))
plt.title('Rescaled Histogram')


# Equalization (Task 2) 
plt.subplot(3,2,5)
plt.imshow(img_eq, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')
plt.subplot(3,2,6)
plt.hist(img_eq.flat, bins=256, range=(0,1))
plt.title('Equalized Histogram')

plt.show()
```

```python
# Task 3: Histogram Matching
# Import required libraries for image processing and visualization
from skimage import data
from skimage.exposure import match_histograms
import matplotlib.pyplot as plt
```

```python
# Load the input image (Chelsea) and the reference image (Rocket)
image = data.chelsea()
reference = data.rocket()
```

```python
# Apply histogram matching
# This adjusts the pixel intensity distribution of the input image
matched = match_histograms(image, reference, channel_axis=-1)
```

```python
# Display the original image, reference image, and matched image
# along with their histograms for comparison

# Figure size
plt.figure(figsize=(12,8))

# Original image
plt.subplot(3,2,1)
plt.imshow(image)
plt.title('Original Image (Chelsea)')
plt.axis('off')

# Original histogram
plt.subplot(3,2,2)
plt.hist(image.ravel(), bins=256, range=(0,255))
plt.title('Original Histogram')

# Reference image
plt.subplot(3,2,3)
plt.imshow(reference)
plt.title('Reference Image (Rocket)')
plt.axis('off')

# Reference histogram
plt.subplot(3,2,4)
plt.hist(reference.ravel(), bins=256, range=(0,255))
plt.title('Reference Histogram')

# Matched image
plt.subplot(3,2,5)
plt.imshow(matched)
plt.title('Matched Image')
plt.axis('off')

# Matched histogram
plt.subplot(3,2,6)
plt.hist(matched.ravel(), bins=256, range=(0,255))
plt.title('Matched Histogram')

plt.tight_layout()
plt.show()
```
